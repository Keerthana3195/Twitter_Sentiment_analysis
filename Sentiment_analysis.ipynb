```python
# import all needed libraries

import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer


nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# read the dataset from a csv file

data = pd.read_csv('C:\\Users\\keert\\Downloads\\SumLab\\Twitter API\\Vertiport.csv')
```

```python
# displaying just the top three rows of the data

data.head(3)
```

```python
# call polarity_scores function to find the polarity scores that the VADER model assigns to each tweet

data['scores'] = data['tweet'].apply(lambda tweet: sia.polarity_scores(tweet))
data[["tweet", "scores"]].head()
```

```python
# fetching only the compound score from all the scores

data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
data[["tweet", "scores", "compound"]].head()
```

```python
# calculating the final sentiment between 'pos' or 'neg' based on the compound score

data['final_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
data[["tweet", "final_score"]].head()
```

```python
# to see the number of 'pos' and 'neg' sentiments in all the tweets
data.final_score.value_counts()
```
